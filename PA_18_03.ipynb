```python
# Operadores aritméticos

# Fazer um programa que o usuário digita 02 números.
# Saída mostrar operações aritméticas.

# Entrada

a = int(input("Digite um número:"))
b = int(input("Digite um número:"))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
resto = a % b
div_int = a // b
potencia = a ** b

# Saída

print("")
print("---------------------OPERADORES ARITMÉTICOS----------------------")
print("")
print("------------------------------SOMA-------------------------------")
print("")
print(f"O resultado da soma de {a} mais {b} é igual à {soma}")
print("-----------------------------------------------------------------")
print("")
print("------------------------------SUBTRAÇÃO--------------------------")
print("")
print(f"O resultado da subtração de {a} menos {b} é igual à {subtracao}")
print("-----------------------------------------------------------------")
print("")
print("------------------------------MULTIPLICAÇÃO----------------------")
print("")
print(f"O resultado da multiplicação de {a} vezes {b} é igual à {multiplicacao}")
print("-----------------------------------------------------------------")
print("")
print("------------------------------DIVISÃO----------------------------")
print("")
print(f"O resultado da divisão de {a} e {b} é igual à {divisao}")
print("-----------------------------------------------------------------")
print("")
print("------------------------------DIVISÃO----------------------------")
print("")
print(f"O resultado da divisão de {a} e {b} é igual à {div_int}")
print("-----------------------------------------------------------------")
print("")
print("------------------------------RESTO------------------------------")
print("")
print(f"O resultado da divisão de {a} e {b} é igual à {resto}")
print("-----------------------------------------------------------------")
print("")
print("------------------------------POTENCIAÇÃO------------------------")
print("")
print(f"O resultado da potenciação de {a} elevado à {b} é igual à {potencia}")
print("-----------------------------------------------------------------")
```

```python
# Fazer um programa que converte em horas em minutos e horas em segundos.

# Entrada

horas = int(input("Digite a quantidade de horas:"))

# Processamento

minutos = horas * 60
segundos = horas * 3600

# Saída

print(f"A quantidade de minutos é igual à {minutos}")
print(f"A quantidade de segundos é igual à {segundos}")
```

```python
# Fazer um programa que converte km em metros.

# Entrada

km = int(input("Digite a quantidade de quilômetros:"))

# Processamento

m = km * 1000

# Saída

print(f"A quantidade de metros é igual à {m}m")
```

```python
# Fazer um programa que converte km em milhas.

# Entrada

km = float(input("Digite a quantidade de quilômetros:"))

# Processamento

milhas = km / 1.609

# Saída

print(f"A quantidade de milhas é igual à {milhas}")

```
