```python
# Escrever um proograma que faz a conversão de temperatura

# Entrada : temperatura em Celsius

temperatura = float(input("Digite a temperatura em Celsius:"))

# Processamento : Converter para Fahrenheit

fahrenheit = (temperatura * 9/5) + 32

# Saída : resultado

print (f"Resultado: {fahrenheit} ")

# Formula : F = (C * 9/5) + 32

```
