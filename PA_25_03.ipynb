```python
# Escrever um programa que receba 04 notas do aluno, calcula a média e mostra se i aluno foi aprovado.


# Entrada

print("Digite suas notas AQUI, valores de 1 à 10 apenas.")
nota1 =  float (input("Digite a nota do primeiro bimestre: "))
nota2 =  float (input("DIgite a nota do segundo bimestre: "))
nota3 =  float (input("Digite a nota do terceiro bimestre: "))
nota4 =  float (input("Digite a nota do quarto bimestre:"))

# Processamento

resultado = (nota1 + nota2 + nota3 + nota4) / 4

if  resultado > 7:
  print (f"A média da suas notas é de {resultado}. Você foi aprovado.")

elif resultado >= 5:

  print(f"A média da suas notas é de {resultado}. Você está de recuperação.")

else:

  print(f" A média da suas notas é de {resultado}. Você foi reprovado.")
```

```python
# Fazer uma calculadora simples com 04 operações

print("----------------------MENU----------------------")
print(" ")
print("Opção - 1 = SOMA")
print("Opção - 2 = SUBTRAÇÃO")
print("Opção - 3 = MULTIPLICAÇÃO")
print("Opção - 4 = DIVISÃO")
print(" ")


opcao = int (input("Digite sua opção: "))


# Entrada, Processamento e Saída


if opcao == 1:
  numero1 = float (input("Você escolheu a opção de SOMA. Digite o primeiro número: "))
  numero2 = float (input("Digite o segundo número:"))

  soma = numero1 + numero2

  print (f" {numero1} + {numero2} = {soma}")



elif opcao == 2:
  numero1 = float (input("Você escolheu a opção de SUBTRAÇÃO. Digite o primeiro número: "))
  numero2 = float (input("Digite o segundo número: "))

  subtracao = numero1 - numero2

  print (f" {numero1} - {numero2} = {subtracao}")



elif opcao == 3:
  numero1 = float (input("Você escolheu a opção de MULTIPLICAÇÃO. Digite o primeiro número: "))
  numero2 = float (input("Digite o segundo número: "))

  multiplicacao = numero1 * numero2

  print (f" {numero1} * {numero2} = {multiplicacao}")



elif opcao == 4:
  numero1 = float (input("Você escolheu a opção de DIVISÃO. Digite o primeiro número: "))
  numero2 = float (input("Digite o segundo número: "))

  divisao = numero1 / numero2

  print (f" {numero1} / {numero2} = {divisao}")



else:
  print("Selecione apenas uma das quatro opções.")




```
